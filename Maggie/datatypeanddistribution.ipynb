```python
df = pd.read_csv("/Users/xuziyan/Desktop/subjects_AND_sampling_metadata_anonymized_full.csv")   
```

```python
print("Shape:", df.shape)
display(df.head())
```

```python
print("Missing values per column:\n")
print(df.isnull().sum().sort_values(ascending=False))
```

```python
print("Data Types:\n")
print(df.dtypes)
```

```python
print("\nSummary Statistics (Numeric):\n")
display(df.describe())

print("\nSummary Statistics (All Columns):\n")
display(df.describe(include="all"))
```

```python
categorical_cols = df.select_dtypes(include=["object", "category"]).columns

print("Categorical Columns:", list(categorical_cols))

for col in categorical_cols:
    print("\n" + "="*60)
    print(f"Value counts for: {col}")
    print("="*60)
    display(df[col].value_counts(dropna=False))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


numeric_cols = df.columns 

for col in numeric_cols:
    s = pd.to_numeric(df[col], errors="coerce")

    s = s.replace([np.inf, -np.inf], np.nan).dropna()

    if s.empty or s.nunique() < 2:
        continue

    plt.figure(figsize=(7,4))
    sns.histplot(s, kde=True, bins=50)  
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
for col in numeric_cols:
    plt.figure(figsize=(7, 2.5))
    sns.boxplot(x=df[col].dropna())
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()
```

```python
print("\nSkewness of numeric variables:\n")
display(df[numeric_cols].skew().sort_values(ascending=False))
```
